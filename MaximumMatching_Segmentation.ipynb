```python
def getDictionary():

    list_Words = []
    f = open('/content/drive/MyDrive/NLP/MaximumMatchingSegmentation/dictionary.txt', "r")
    text = f.readlines()
    for i in text:
        list_Words.append(i.replace("\n",""))
    # print((len(list_Words)))
    return list_Words
```

```python
getDictionary()
```

```python
def maximumMatching(sentence, maxlen):
  
    VNdictionary = getDictionary()
    my_list = []
    list_Words = sentence.split()
    len_now = len(list_Words)
    i = 0
    maxl = maxlen
    while i < len_now: #13
        word = '_'.join(list_Words[i:maxlen + i])
        while word not in VNdictionary:
            if maxl > 1:
                word = '_'.join(list_Words[i:i + maxl])
            elif maxl == 1:
                word = list_Words[i]
            elif maxl == 0:
                break
            maxl -= 1
        if maxl > 0:
            i += maxl
        i += 1
        my_list.append(word)
        # print(my_list)
        maxl = min(len_now - i, maxlen)

    line = ' '.join(my_list)
    return line
```

```python
def WordSegmentation(path, maxlen=4):
    with open(path, 'r', encoding='utf-8') as f:
        text = f.readlines()
    new_text=[]
    for i in text:
      new_text.append(i.replace(","," ,").replace("."," .").replace("?"," ?").replace('"',' " ').replace("\n",""))
    # print(new_text)
    with open('/content/drive/MyDrive/NLP/MaximumMatchingSegmentation/output_maximum_tokenize.txt', 'w', encoding='utf-8') as f:
        for i in new_text:
            sample = maximumMatching(i,maxlen).split()
            # print(sample)
            for j in sample:
                f.write('{}\n'.format(j))
            # f.write('.\n')
            f.write('\n')
```
